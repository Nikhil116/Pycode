```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
html = requests.get(url).content
```

```python
df_list = pd.read_html(html)
```

```python
df = df_list[0]
print(df.head())
print("\n")
print(df.shape)
print(type(df))
```

```python
df.replace("Not assigned", np.nan, inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
nan_rows = df[df['Borough'].isnull()]
nan_rows
```

```python
nan_rows.shape
```

```python
nan_rows = df[df['Neighborhood'].isnull()]
nan_rows.shape
```

```python
df.dropna(axis=0, inplace=True)
df.head()
```

```python
print(df.shape)
print(type(df))
print("\n")
print(df.isnull().sum())
```
